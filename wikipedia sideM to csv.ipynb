```python
import pprint
import re

with open('sidem.txt', "r", encoding="utf-8") as f:
    s = f.read()
s = s.replace('\u3000', '、')
lines = s.splitlines()
songs = []
for i in range(len(lines)):
    line = lines[i]
    if line.startswith('作詞'):
        result = re.match('作詞：(.*)、作曲・編曲：(.*)$', line)
        result2 = re.match('作詞・作曲：(.*)、編曲：(.*)$', line)
        result3 = re.match('作詞：(.*)、作曲：(.*)、編曲：(.*)$', line)
        if result:
            songs.append({'曲名':lines[i-1], '作詞': result.group(1), '作曲': result.group(2), '編曲': result.group(2)})
        elif result2:
            songs.append({'曲名':lines[i-1], '作詞': result2.group(1), '作曲': result2.group(1), '編曲': result2.group(2)})
        elif result3:
            songs.append({'曲名':lines[i-1], '作詞': result3.group(1), '作曲': result3.group(2), '編曲': result3.group(3)})
        else:
            songs.append({'曲名':lines[i-1], 'information': line})

for i in range(len(songs)):
    song = songs[i]
    if 'information' in song:
        print(song)
pprint.pprint(songs)
print(len(songs))
```

```python
import csv

with open('sidem.csv', 'w', encoding='utf-8') as f:
    writer = csv.DictWriter(f, ['曲名', '作詞', '作曲', '編曲'],lineterminator='\n')
    writer.writeheader()
    writer.writerows(songs)
```

```python
import csv
import pprint

songs = []

with open('songs.csv', 'r',encoding="utf-8") as f:
    # DictReaderと共にHeaderを渡すことで辞書形式で返す。
    # 取得したデータを1行ずつ出力。
    reader = csv.DictReader(f)
    for row in reader:
        songs.append(row)
pprint.pprint(songs)
```

```python
#名寄せのために、所属を削除
import re
import json

songs_text = json.dumps(songs, indent=4, ensure_ascii=False)
songs_text = re.sub('[（\(]Arte Refact[）\)]', '',songs_text)
songs_text = re.sub('[（\(]Nostalgic Orchestra[）\)]', '',songs_text)
songs_text = re.sub('[（\(]Void_Chords[）\)]', '',songs_text)
songs_text = re.sub('[（\(]Dream Monster[）\)]', '',songs_text)
songs_text = re.sub('[（\(]fhana[）\)]', '',songs_text)
songs_text = re.sub('[（\(]onetrap[）\)]', '',songs_text)
songs_text = re.sub('[（\(]MONACA[）\)]', '',songs_text)
songs_text = re.sub('[（\(]MONACA[）\)]', '',songs_text)
songs_text = re.sub('NBSI[\(（]([^\)）]*)[\)）]', '\\1',songs_text)
songs_text = re.sub('BNSI[\(（]([^\)）]*)[\)）]', '\\1',songs_text)
songs_text = re.sub('NBGI[\(（]([^\)）]*)[\)）]', '\\1',songs_text)
songs_text = re.sub('BNGI[\(（]([^\)）]*)[\)）]', '\\1',songs_text)
songs_text = re.sub('BNEI[\(（]([^\)）]*)[\)）]', '\\1',songs_text)
songs = json.loads(songs_text)
```

```python
import collections

for song in songs:
    artists = []
    artists.append(song['作詞'])
    artists.append(song['作曲'])
    #アーティスト名にカンマが入っていたら分離
    for i in range(len(artists)):
        artist = artists[i]
        sep = artist.find('、')
        if sep>0:
            splitted = artist.split('、')
            artists[i] = ''
            artists.extend(splitted)
    
    #個数を数える
    c = collections.Counter(artists)
    c.pop('', None)
    song['score'] = dict(c)
    scores = []
    for key in c.keys():
        scores.append({'artist': key, 'value':c[key]})
    song['scores'] = scores
    pprint.pprint(song)
```

```python
#保存
import json
import codecs
#hoge = codecs.open([JSONfile], 'w', 'utf-8')
#json.dump([JSONfile], hoge, ensure_ascii=False)

with codecs.open('imas_songs.json', 'w','utf-8') as f:
    
    d = json.dump(songs, f, indent=4, ensure_ascii=False)
    pprint.pprint(d)
```
